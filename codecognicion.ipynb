```python
import pandas as pd
from scipy.stats import shapiro
from itertools import product

# Supongamos que tienes un DataFrame llamado 'df' con todas las variables mencionadas

# Lista de variables que deben ser verificadas
variables = [
    'SEG INSTRUC DIR', 'DER IZQ DIR', 'LATERALIDAD', 'CONTRALATERALIDAD',
    'CAN LETRAS DIR', 'DIG PROG DIR', 'DIG REGR DIR', 'MATRICES DIR',
    'SIMILITUDES DIR', 'ORIENTAC LÍNEAS', 'MENTAL FOLDING', 'ANIMALES EN FILA',
    'SCALING TEST TOTAL', 'SCALINT TEST 11', 'SCALINT TEST 12', 'SCALINT TEST 14',
    'SCALINT TEST mm TOTAL girado', 'CWT NUM MOV 0', 'CWT NUM MOV 02',
    'CWT MOV INCORR 0', 'CWT MOV INCORR 02', 'LATENCIAS TOTAL CERO GRADOS',
    'TIEMPO EJECUC 0grados TOTAL', 'INICIO CORR 0', 'INICIO CORR 02',
    'LADO CORREC 0', 'LADO CORREC 02', 'CWT NUM MOV 90', 'CWT NUM MOV 902',
    'CWT MOV INCORR 90', 'CWT MOV INCORR 902', 'LATENCIA 90 TOTAL', 'TIEMPO 90 TOTAL',
    'INICIO CORR 90', 'INICIO CORR 902', 'LADO CORREC 90', 'LADO CORREC 902',
    'ERROR EGO 90', 'ERROR EGO 902', 'CWT NUM MOV 180', 'CWT NUM MOV 1802',
    'CWT MOV INCORR 180', 'CWT MOV INCORR 1802', 'LATENCIA 180 TOTAL', 'TIEMPO 180 TOTAL',
    'INICIO CORR 180', 'INICIO CORR 1802', 'LADO CORREC 180', 'LADO CORREC 1802',
    'ERROR EGO 180', 'ERROR EGO 1802', 'COORDENADAS', 'SIN COORDENADAS', 'HORIZONTALES',
    'VERTICALES', 'CENTRAL', 'ANTERIOR', 'POSTERIOR', 'OTROS', 'DEICTICOS'
]

# Variables que definen los grupos
grupos = ['POBEDAD']

# Combinaciones de variables de grupo
combinaciones = list(product(grupos, variables))

# Realizar el test de Shapiro-Wilk para cada combinación
resultados = []
for grupo, variable in combinaciones:
    grupo_variable = df[[grupo, variable]].dropna()
    for nombre_grupo, datos_grupo in grupo_variable.groupby(grupo):
        _, p_value = shapiro(datos_grupo[variable])
        resultados.append({
            'Grupo': nombre_grupo,
            'Variable': variable,
            'p-valor': p_value,
            'Normalidad': 'Normal' if p_value > 0.05 else 'No normal'
        })

# Crear un DataFrame con los resultados
resultados_df = pd.DataFrame(resultados)

# Guardar los resultados en un archivo CSV
resultados_df.to_csv('resultados_normalidad.csv', index=False)

# Puedes imprimir los resultados si lo deseas
print(resultados_df)

```
